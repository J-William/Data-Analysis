```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Data file path
fifa_filepath = "datasets/fifa/fifa.csv"
# Read
fifa_data = pd.read_csv(fifa_filepath, index_col = "Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
fifa_data.info()
```

```python
#Set width and height of figure
plt.figure(figsize=(16,6))

# Line chart showing fifa rankings over time
sns.lineplot(data=fifa_data)
```

```python
spotify_filepath = "datasets/spotify/spotify.csv"
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

```python
# Line Chart daily global streams of each song in millions
plt.figure(figsize=(14,6))
sns.lineplot(data=spotify_data)
```

```python
# List the columns
list(spotify_data.columns)
```

```python
plt.figure(figsize=(14,6))
# Title
plt.title("daily global streams of popular songs in 2017-2018")

sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

plt.xlabel("Date")
```

```python
flight_filepath = "datasets/flightdelays/flight_delays.csv"
flight_data= pd.read_csv(flight_filepath, index_col="Month")
```

```python
flight_data.info()
```

```python
flight_data
```

```python
plt.figure(figsize=(10,6))
plt.title("average arrival delay for spirit airlines flights, by month")
sns.barplot(x=flight_data.index, y=flight_data['NK'])

plt.ylabel("arrival dealy in minutes")
```

```python
# Heat Map
plt.figure(figsize=(14,7))

plt.title("average arrival delay for each airline by month")

sns.heatmap(data=flight_data, annot=True)

plt.xlabel("Airline")
```
